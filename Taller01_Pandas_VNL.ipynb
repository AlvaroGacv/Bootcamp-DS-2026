```python
# ============================================================
# TALLER 01 - MANIPULACIÓN Y ANÁLISIS DE DATOS CON PANDAS
# Dataset: Volleyball Nations League (VNL) 2025
# Autor: [Alvaro Gallego Acevedo]
# Fecha: Abril 2026
# ============================================================

# Importamos las librerías necesarias para el análisis
import pandas as pd        # Para manipulación y análisis de datos
import numpy as np         # Para operaciones matemáticas
import matplotlib.pyplot as plt  # Para visualización de datos

print("✅ Librerías importadas correctamente")
print(f"Versión de Pandas: {pd.__version__}")
print(f"Versión de NumPy: {np.__version__}")
```

```python
# ============================================================
# PASO 1: CARGA Y EXPLORACIÓN DEL DATASET
# ============================================================

# Cargamos el dataset de la VNL 2025 desde el archivo CSV
# Usamos encoding='latin1' porque el archivo contiene caracteres especiales
df = pd.read_csv(r"C:\Users\alvar\Downloads\archive\VNL2025.csv", encoding="latin1")

print("✅ Dataset cargado correctamente")
print(f"El dataset contiene {df.shape[0]} jugadores y {df.shape[1]} columnas\n")

# Mostramos las primeras 5 filas para entender la estructura del dataset
print("=== PRIMERAS 5 FILAS DEL DATASET ===")
display(df.head())
```

```python
# Información general del dataset: tipos de datos y valores nulos
print("=== INFORMACIÓN GENERAL DEL DATASET ===")
df.info()

# Estadísticas descriptivas: media, desviación estándar, mínimo, máximo, etc.
print("\n=== ESTADÍSTICAS DESCRIPTIVAS ===")
display(df.describe())
```

```python
# ============================================================
# PASO 2: LIMPIEZA Y TRANSFORMACIÓN DE DATOS
# ============================================================

# Verificamos si hay valores nulos en cada columna
print("=== VALORES NULOS POR COLUMNA ===")
print(df.isnull().sum())
print("\n✅ No se encontraron valores nulos, el dataset está limpio")

# Creamos una nueva columna con la eficiencia de ataque
# Fórmula: (Puntos de Ataque - Errores de Ataque) / Intentos de Ataque
# Usamos replace(0,1) para evitar división por cero
df["Attack Efficiency"] = round(
    (df["Attack Points"] - df["Attack Errors"]) / df["Attack Attempts"].replace(0, 1), 2
)

# Creamos una columna con el total de puntos anotados por jugador
# Suma de: puntos de ataque + puntos de bloqueo + puntos de saque
df["Total Points"] = df["Attack Points"] + df["Block Points"] + df["Serve Points"]

print("\n✅ Nuevas columnas creadas:")
print("   - 'Attack Efficiency': mide qué tan eficiente es el jugador atacando")
print("   - 'Total Points': suma total de puntos anotados por el jugador")
print("\n=== VISTA PREVIA CON NUEVAS COLUMNAS ===")
display(df[["Name", "Team", "Total Points", "Attack Efficiency"]].head(10))
```

```python
# ============================================================
# PASO 3: FILTROS Y AGRUPACIONES
# ============================================================

# Filtramos y ordenamos el Top 10 jugadores con más puntos totales
print("=== TOP 10 JUGADORES POR PUNTOS TOTALES ===")
top10 = df[["Name", "Team", "Total Points"]].sort_values(
    "Total Points", ascending=False
).head(10)
display(top10)

# Agrupamos por equipo y calculamos el promedio de puntos
# Esto nos permite comparar el rendimiento general de cada selección
print("\n=== PROMEDIO DE PUNTOS POR EQUIPO ===")
por_equipo = df.groupby("Team")["Total Points"].mean().round(2).sort_values(ascending=False)
display(por_equipo)

# Filtramos jugadores con más de 100 puntos totales (jugadores destacados)
print("\n=== JUGADORES DESTACADOS (más de 100 puntos) ===")
destacados = df[df["Total Points"] > 100][["Name", "Team", "Total Points"]].sort_values(
    "Total Points", ascending=False
)
display(destacados)
```

```python
# ============================================================
# PASO 4: VISUALIZACIÓN DE DATOS
# ============================================================

# Gráfica 1: Top 10 jugadores por puntos totales
plt.figure(figsize=(12, 6))
plt.barh(
    top10["Name"] + " (" + top10["Team"] + ")",
    top10["Total Points"],
    color="steelblue"
)
plt.xlabel("Total de Puntos")
plt.title("🏐 Top 10 Jugadores VNL 2025 - Puntos Totales")
plt.gca().invert_yaxis()  # Invertimos para que el primero quede arriba
plt.tight_layout()
plt.show()

# Gráfica 2: Promedio de puntos por equipo
plt.figure(figsize=(12, 6))
por_equipo.plot(kind="bar", color="coral")
plt.xlabel("Equipo")
plt.ylabel("Promedio de Puntos")
plt.title("🏐 Promedio de Puntos por Equipo - VNL 2025")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfica 3: Distribución de puntos totales
plt.figure(figsize=(10, 5))
plt.hist(df["Total Points"], bins=20, color="mediumseagreen", edgecolor="black")
plt.xlabel("Total de Puntos")
plt.ylabel("Número de Jugadores")
plt.title("🏐 Distribución de Puntos Totales - VNL 2025")
plt.tight_layout()
plt.show()
```
